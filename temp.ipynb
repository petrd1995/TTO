```python
import numpy as np
import plotly.express as px
```

```python
class buckle():

    def __init__(self, r, E, l, Fratio) -> None:
        self.r = r
        self.E = E
        self.l = l
        self.Fratio = Fratio

        self.A = np.pi*self.r**2  # mm**2
        self.J = np.pi*self.r**4/64  # mm**4
        self.Fcrit = np.pi**2*self.E*self.J/self.l**2  # N
        self.F = self.Fcrit*self.Fratio  # N
        self.α = self.F*self.l**2/(self.E*self.J)
        self.ξ = (self.α/2)**2-(np.pi**2 - self.α/2)**2
        self.c = self.E*self.J/self.l**4*self.ξ

    
```

```python
case1 = buckle(5, 2.1e5, np.arange(1, 25, 0.1), 2)
```

```python
fig = px.line(x=case1.l, y=case1.c, log_y=True, labels={
              'x': 'délka prutu [mm]', 'y': 'E2 [MPa]'}, title='Závislost délky prutu na navrhovaném E2')
fig.show()
```
